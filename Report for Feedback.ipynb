```python
import os
import re
import subprocess
import sys
import sysconfig

PROJECT_DIR = "Report for Feedback.ipynb"  # change if needed
OUTPUT_FILE = "requirements.txt"

# Get list of standard library modules to exclude
stdlib_path = sysconfig.get_paths()["stdlib"]
stdlib_modules = set()
for root, _, files in os.walk(stdlib_path):
    for file in files:
        if file.endswith(".ipynb") and file != "__init__.py":
            stdlib_modules.add(file[:-3])

# Also exclude built-in modules from sys.builtin_module_names
stdlib_modules.update(sys.builtin_module_names)

# Regex to capture "import x" or "from x import y"
import_pattern = re.compile(r'^\s*(?:from|import)\s+([a-zA-Z0-9_\.]+)')

found_packages = set()

# Walk all Python files
for root, _, files in os.walk(PROJECT_DIR):
    for file in files:
        if file.endswith(".py"):
            with open(os.path.join(root, file), "r", encoding="utf-8") as f:
                for line in f:
                    match = import_pattern.match(line)
                    if match:
                        pkg = match.group(1).split('.')[0]
                        if pkg not in stdlib_modules:
                            found_packages.add(pkg)

# Get installed packages and versions from pip freeze
installed_packages = subprocess.check_output([sys.executable, "-m", "pip", "freeze"], text=True).splitlines()
pkg_versions = {p.split("==")[0].lower(): p for p in installed_packages}

# Match found packages with installed ones
requirements = []
for pkg in sorted(found_packages):
    pkg_lower = pkg.lower()
    if pkg_lower in pkg_versions:
        requirements.append(pkg_versions[pkg_lower])
    else:
        requirements.append(pkg)  # no version found

# Write requirements.txt
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    f.write("\n".join(requirements))

print(f"✅ Extracted {len(requirements)} third-party packages to {OUTPUT_FILE}")
```

```python
import re
import sys
import sysconfig
import subprocess

# اسم الملف اللي شغال عليه
PY_FILE = "dashboard.py"  # غيّره لو اسم ملفك مختلف
OUTPUT_FILE = "requirements.txt"

# نجيب مكتبات الـ Standard Library
stdlib_path = sysconfig.get_paths()["stdlib"]
stdlib_modules = set()
import os
for root, _, files in os.walk(stdlib_path):
    for file in files:
        if file.endswith(".py") and file != "__init__.py":
            stdlib_modules.add(file[:-3])
stdlib_modules.update(sys.builtin_module_names)

# Regex علشان نمسك import و from ... import ...
import_pattern = re.compile(r'^\s*(?:from|import)\s+([a-zA-Z0-9_\.]+)')

found_packages = set()

# قراءة الملف واستخراج المكتبات
with open(PY_FILE, "r", encoding="utf-8") as f:
    for line in f:
        match = import_pattern.match(line)
        if match:
            pkg = match.group(1).split('.')[0]
            if pkg not in stdlib_modules:
                found_packages.add(pkg)

# نجيب الإصدارات من pip freeze
installed_packages = subprocess.check_output([sys.executable, "-m", "pip", "freeze"], text=True).splitlines()
pkg_versions = {p.split("==")[0].lower(): p for p in installed_packages}

# تجهيز قائمة المكتبات
requirements = []
for pkg in sorted(found_packages):
    pkg_lower = pkg.lower()
    if pkg_lower in pkg_versions:
        requirements.append(pkg_versions[pkg_lower])
    else:
        requirements.append(pkg)

# كتابة الملف
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    f.write("\n".join(requirements))

print(f"✅ تم استخراج {len(requirements)} مكتبة من {PY_FILE} وكتابتها في {OUTPUT_FILE}")
```
