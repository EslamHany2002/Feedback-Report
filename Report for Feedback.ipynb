```python
df = pd.read_csv('Clean Feedback.csv')
df
```

```python
df['Techincal / Operation Solved'] = df['Techincal / Operation Solved'].str.lower().str.strip()


df_solved = df[df['Techincal / Operation Solved'] == 'solved']
solved_counts = df_solved.groupby('Select your assigned group | اختر المجموعة المخصصة لك').size().reset_index(name='count')

plt.figure(figsize=(10,5))
sns.barplot(data=solved_counts, x='Select your assigned group | اختر المجموعة المخصصة لك', y='count', palette='Greens')
plt.title('Number of Solved per Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_follow = df[df['Techincal / Operation Solved'] == 'follow up']
follow_counts = df_follow.groupby('Select your assigned group | اختر المجموعة المخصصة لك').size().reset_index(name='count')

plt.figure(figsize=(10,5))
sns.barplot(data=follow_counts, x='Select your assigned group | اختر المجموعة المخصصة لك', y='count', palette='Oranges')
plt.title('Number of Follow Up per Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_follow = df[df['Techincal / Operation Solved'] == 'not solved']
follow_counts = df_follow.groupby('Select your assigned group | اختر المجموعة المخصصة لك').size().reset_index(name='count')

plt.figure(figsize=(10,5))
sns.barplot(data=follow_counts, x='Select your assigned group | اختر المجموعة المخصصة لك', y='count', palette='Reds')
plt.title('Number of Not Solved per Group')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df['Techincal / Operation Solved'] = df['Techincal / Operation Solved'].str.lower().str.strip()

# تصفية على solved
df_solved = df[df['Techincal / Operation Solved'] == 'solved']

# تنظيف عمود اللينكات
df_solved['Evidence Attachment'] = df_solved['Evidence Attachment'].astype(str).str.strip()

# تحديد إذا فيه لينك أو فاضي
df_solved['Has Link'] = df_solved['Evidence Attachment'].apply(
    lambda x: 'Link' if x != '' and x.lower() != 'nan' else 'Empty'
)

# حساب القيم
link_counts = df_solved['Has Link'].value_counts()


# رسم Bar Chart
plt.figure(figsize=(6,4))
plt.bar(link_counts.index, link_counts.values, color=['#4CAF50','#FF5722'])
plt.title('Evidence Attachment')
plt.ylabel('عدد الصفوف')
plt.show()
```
